# Çapraz Tablolarda Ki-Kare ($\chi^2$) Testi 

**Ki-Kare Dağılımına Dayanan Hipotez Testi:**

+ Ki-Kare dağılımına dayanan hipotez testleri, kitle dağılımı hakkında herhangi bir bilgi ya da varsayımın olmadığı durumlarda kullanılır. Yani, kategorik verilerin bağımısızlığını veya homojenliğini test etmeye yarar.

**Çapraz Tablolar:**

+ İki farklı nitel özellik (değişken) olduğunda, verilen nasıl gösteren tablolardır. Bu tablolara aynı zamanda `iki yönlü tablo`, `olumsallık tablosu` da denir.

Bu dosyada formülleri göstermeyeceğim. Direkt olarak `scipy` ile çapraz tablolarda nasıl Ki-Kare testi yapıldığını bir örnek üzerinde göstereceğim.

## Örnek

Covid-19'un obezite ile ilişki olup olmadığını aşağıdaki 2x2 çapraz tablosuna bakarak inceleyelim. (a = 0.05)



## Çözüm

Hipotezi Kuralım: 

$H_0:$ Obezite ve Covid-19 değişkenleri birbirlerinden bağımsızdır.

$H_1:$ Obezite ve Covid-19 değişkenleri birbirlerinden bağımsız değildir.

```python
import scipy.stats 
import numpy as np

negative = [12, 58]
positive = [79, 1442]
data = np.array([negative, positive])

#Bağımsızlık Testi
chi2, p, dof, expected = scipy.stats.chi2_contingency(data)
print(f"Ki-Kare Test İstatistiği: {chi2} \n P Değeri: {p} \n Serbestlik Derecesi: {dof} \n Beklenen Frekanslar: {expected}")
```

+ P değeri a'dan büyük olduğu için $H_0$ reddedilemez. Yani obezite ve Covid - 19 arasında bir ilişki vardır.
