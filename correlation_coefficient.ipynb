# Korelasyon Katsayısı 

İki değişken arasındaki ilişkinin araştırılması sonucu elde edilen katsayıya **`Korelasyon Katsayısı`** denir. Bu ilişkinin araştırılmasına da **`Korelasyon Çözümlemesi`** adı verilir ve bu çözümleme sonucu korelasyon katsayısı elde edilir. Kitle korelasyon katsayısı **`P`** ile, örneklem korelasyon katsayısı **`r`** ile gösterilir. 

Korelasyon katsayısı, değişkenler arasındaki ilişkinin derecesini ve yönünü belirtir. Yani bir değişkenin aldığı değerler değişirken, bir başka değişkenin aldığı değerler de buna bağlı olarak sistematik bir şekilde değişim gösteriyorsa bu iki değişken arasında bir ilişki (Korelasyon) vardır. Korelasyon katsayısını belirlemek için çeşitle ölçütler geliştirilmiştir. Bunlar: 

+ Pearson'un Korelasyon Katsayısı
+ Spearman'ın Sıra (Rank) Korelasyon Katsayısı 
+ Kendall'ın Tau ($t$) Korelasyon Katsayısı'dır.

Korelasyon katsayısı **`Basit Korelasyon Katsayısı`** ve **`Kısmi Korelasyon Katsayısı`** olarak ikiye ayrılır. Basit korelasyon katsayısı; yalnızca iki değişken arasındaki ilişkiyi, Kısmi korelasyon katsayısı; ikiden çok değişken arasındaki ilişkiyi gösterir. 

Korelasyon katsayısının hesaplanmasında **`Kovaryans`** kullanılır. Kovaryans, iki raslantı değişkeninin birlikte nasıl değişim gösterdiğini ifade eder. Kovaryansın hesaplanışı: 

$\mu_X$ = X raslantı değişkeninin ortalaması

$\mu_Y$ = Y raslantı değişkeninin ortalaması olmak üzere:

Cov(X, Y) = $E[(X - \mu_X).(Y - \mu_Y)]$ olarak hesaplanır. Eğer bu hesap sonucu elde edilen Cov(X, Y) değeri:


+ Cov(X, Y) > 0 ise X ve Y aynı yönde artış ya da azalış gösteriyor demektir. 
+ Cov(X, Y) < 0 ise X ve Y ters yönde artış ya da azalış gösteriyor demektir.
+ Cov(X, Y) = 0 ise X ve Y arasında bir ilişki yoktur. Yani bağımsızlar demektir.

Kovaryans ilişikinin sadece yönünü belirtir, derecesi hakkında bilgi vermez. Bu yüzden de yukarıda belirtilen ölçütleri kullanırız. `Bu ölçütler, standartlaştırılmış değerler arasındaki kovaryanstır.`

Korelasyon katsayısı kabaca şu adımlarla bulunur: 

+ 1-) İlk olarak arasında ilişkinin olup olmadığı incelenecek örneklemlerin Normal dağılıma sahip olup olmadığı `Shapiro Wilk Normallik testi` ile test edilir. 

+ 2-) Eğer ilişkinin inceleneceği iki örneklem de Normal dağılıma sahipse `Pearson'un Korelasyon Katsayısı` ile, Normal dağılıma sahip değilse `Spearman'ın Sıra (Rank) Korelasyon Katsayısı` ile korelasyon katsayısı hesaplanır.

+ 3-) Daha sonra bulunan bu korelasyon katsayısının anlamlılığını test ederiz. Bu anlamlılık testi aslında örneklem korelasyon katsayısını bulduğumuzda, kitle korelasyon katsayısı için yaptığımız bir hipotez testidir. Bu test de ikiye ayrılır:

    + $H_0: P = 0$
    
    + $H_0: P \neq 0$ her hipotez, farklı bir durumu ifade eder. Bunları aşağıdaki kısımlarda göreceğiz.

## Pearson'un Korelasyon Katsayısı

X ve Y raslantı değişkenleri normal dağılıma sahip olduğunda aralarındaki ilişki doğrusal olur. Pearson'un korelasyon katsayısı da doğrusal ilişkiye sahip değişkenler arasındaki korelasyon katsayısını hesaplar. Eğer X ve Y raslantı değişkenleri arasındaki ilişki doğrusal değilse yani, X ve Y raslantı değişkenleri normal dağılıma sahip değilse, Pearsoun'un korelasyon katsayısı gerçek durumu yansıtamaz. İşte böyle durumlarda da Spearman'ın Sıra (Rank) korelasyon katsayısı kullanılır. Dolayısıyla korelasyon katsayısını hesaplamadan önce **`Shapiro Wilk Normallik Testi`** yapılmalıdır. 

Pearson'un korelasyon katsayısı formülü:

+ Kitle için: $P$ = $\frac{E[(X - \mu_X).(Y - \mu_Y)]} {\sqrt{E(X - \mu_X)^2 . E(Y - \mu_Y)^2}}$ = $\frac{Cov(X, Y)}{\sqrt{V(X). V(Y)}}$

+ Örneklem için: $r$ = $\frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2 \sum_{i=1}^{n} (y_i - \bar{y})^2}}$

Bu formüller sonucunda elde edilen sayılar $[-1, 1]$ aralığında olacaktır. Korelasyon katsayısı -1'e yaklaştıkça ilişki negatif yönlü, 1'e yaklaştıkça pozitif yönlü olur ve 0'dan uzaklaştıkça ilişkinin derecesi artar.

## Spearman'ın Sıra (Rank) Korelasyon Katsayısı 

X ve Y raslantı değişkenleri normal dağılıma sahip olmadığında kullanılır. Pearson korelasyon katsayısının **`Parametrik Olmayan`** karşılığıdır. Bu iki ölçütün arasındaki en önemli fark: Pearson, gözlemlerin ham değerlerini yani direkt olarak kendisini kullanarak hesap yaparken Spearman, gözlemlerin sıra değerleri ile hesap yapar.

Spearman'ın sıra korelasyon katsayısını bulmak için öncelikle X ve Y raslantı değişkenlerinin gözlem değerleri ayrı ayrı küçükten büyüğe sıralanır. X raslantı değişkeninin sıra numaraları $R_i$ ve Y raslantı değişkenin sıra numaları $S_i$ olmak üzere $d_i = R_i - S_i$ olarak hesaplanır. Spearman'ın sıra korelasyon katsayısının formülü ise:

+ $r_s$ = $1 - \frac{6. \sum_{i=1}^{n} d_i^2}{n.(n^2-1)}$ şeklinde hesaplanır.

Bu formüller sonucunda elde edilen sayılar $[-1, 1]$ aralığında olacaktır. Korelasyon katsayısı -1'e yaklaştıkça ilişki negatif yönlü, 1'e yaklaştıkça pozitif yönlü olur ve 0'dan uzaklaştıkça ilişkinin derecesi artar.

## Korelasyon Katsayısının Anlamlılık Testi 

Kitleye ilişkin korelasyon katsayısıyla ilgili hipotez testlerinin yapılabilmesi için korelasyon katsayısının örnekleme dağılımı bilinmelidir. İlk olarak burada `örnekleme dağılımı` nedir bunu görelim:



Korelasyon katsayısı sıfır olan bir kitleden rastgele seçilen örneklemlerin korelasyon katsayılarının örnekleme dağılımı, ortalaması sıfır olan bir Normal dağılım gösterir.

Eğer kitle korelasyon katsayısı sıfırdan farklı ise, kitleden çekilen örneklemlerin korelasyon katsayılarının örnekleme dağılımı Normal dağılım göstermez. Yani dağılım çarpıktır. Bu durumda örneklem korelasyon katsayısı (r)'ye **`Fisher'ın Z Dönüşümü`** uygulanır ve dağılım Normal dağılıma yakınlık gösterir.

### $P = 0$'ın Anlamlılık Testi 

Kitle korelasyon katsayısı P = 0 olan bir kitleden seçilen örneklemlerin korelasyon katsayılarının örnekleme dağılımı, ortalaması sıfır olan Normal dağılım gösterir. Böylece: 

+ r'nin beklenen değeri: $E(r) = 0$ ve

+ Standart hatası: $S_r = \sqrt{\frac{1 - r^2}{n - 2}}$ olur.

+ Hipotezi: $H_0: P = 0$ ve $H_1: P \neq 0$ ya da $H_1: P \leq 0$ ya da $H_1: P \geq 0$ şeklinde kurulur.

+ Standart hata örneklemden tahmin edildiği için test istatistiği de: $t_H = \frac{r}{S_r} = \frac{r}{\sqrt{\frac{1 - r^2}{n-2}}}$ olur ve serbestlik derecesi $n - 2$ olan $t$ dağılımı gösterir.


### $P = P_0$'ın Anlamlılık Testi 

Kitle korelasyon katsayısı sıfırdan farklı ise kitleden seçilen örneklemlerin korelasyon katsayılarının örnekleme dağılımı Normal dağılıma sahip değildir. Bu durumda anlamlılık testi yapmak için örneklem korelasyon katsayısı (r)'ye `Fisher Dönüşümü` uygulanır. Bu dönüşüm sonucunda `r`'nin örnekleme dağılımı, $n \geq 30$ olan durumlarda Normal dağılıma yakınsar. Fisher dönüşümü:

+ $Z_r = \frac{1}{2} \ln\left(\frac{1+r}{1-r}\right)$ şeklindedir. Fisher, Z raslantı değişkeninin: 

    + Ortalamasının: $Z_p = \frac{1}{2} \ln\left(\frac{1+p}{1-p}\right)$
    
    + Standart sapmasının: $\sigma_Z = \frac{1}{\sqrt{n-3}}$ olduğunu ve $Z_r$'nin yaklaşık olarak normal dağıldığını ispatlamıştır.
    
Z raslantı değişkeni Normal dağılıma yakınsadığı için Normal dağılımın özelliklerinden faydalanıp bu dağılımı standartlaştırabiliriz. Böylece de test istatistiğini elde etmiş oluruz.

+ Z raslantı değişkeninin standartlaştırılmış hali = Test İstatistiği: $Z_H$ = $\frac{Z_r - Z_p}{\sigma_Z} = [\frac{1}{2} \ln\left(\frac{1+r}{1-r}\right) - \frac{1}{2} \ln\left(\frac{1+p}{1-p}\right)].\sqrt{n-3}$

    + Burada: $Z_r = \bar{x}$ ve $Z_p = \mu$ ve $\sigma_Z = \sigma$ gibi olmuş oluyor. (SADECE BİR BENZETME!! BU DEĞERLER DEMEK DEĞİL!!)

## Örnek

`winequality-red.csv` dosyasındaki `fixed acidity` ve `volatile acidity` arasındaki korelasyonu inceleyelim. İlk önce Shapiro Wilk testi ile normalliğini test edeceğiz. Ardından çıkan sonuca göre bir korelasyon katsayısı ölçütü ile korelasyon katsayısını bulacağız. Son olarak da anlamlılık testi yapacağız.

```python
import os

os.chdir("C:\\Users\\ozero\\ÖNEMLİ ÇALIŞMALAR")
```

### Veriyi içe aktaralım

```python
import pandas as pd 
df = pd.read_csv("winequality-red.csv")
df.head(5)
```

### Shapiro Wilk Normallik Testini Yapalım

```python
import scipy.stats
print(scipy.stats.shapiro(df["fixed acidity"]))
```

```python
print(scipy.stats.shapiro(df["volatile acidity"]))
```

+ İki grubun da P değerleri a'dan (0.05)'ten büyük olduğu için normal dağıldıklarını söyleyebiliriz. Normal dağılım gösterdikleri için Pearson'un korelasyon katsayısı ölçütünü kullanacağız.

### Pearson'un korelasyon katsayısı

```python
r, p = scipy.stats.pearsonr(df["fixed acidity"], df["volatile acidity"])
print("Korelasyon katsayısı:",r, "/" ,"P değeri",p)
```

+ Korelasyon katsayını -0.25 olarak bulduk. Fixed acidity ile volatile acidity arasında negatif yönde zayıf bir ilişki olduğunu söyleyebiliriz. Son olarak kitle korelasyon katsayısı için hipotez testi yapalım. Yani anlamlılık testini uygulayalım. 

### P = 0'ın anlamlılık testi

$H_0: P = 0$

$H_1: P \neq 0$ hipotezini a = 0.05 düzeyinde test edelim.

```python
#Test istatistiğinin hesabı
import numpy as np
n = len(df["volatile acidity"]) #fixed acidity de olabilirdi. ikisi de aynı gözlem değerine sahip 
test_stat = r / np.sqrt((1-r**2)/(n-2))
print(test_stat)
```

```python
#Tablo değerinin hesabı
a = 0.05
t_table_upper = scipy.stats.t.ppf(1-(a/2), df = n-2)
t_table_lower = scipy.stats.t.ppf(a/2, df = n-2)
print("Alt sınır:",t_table_lower, "/", "Üst sınır:",t_table_upper)
```

+ Test istatistiği -10.5 ret bölgesine düştüğü için $H_0$ hipotezi reddedilir. Yani bu korelasyon katsayısı istatistiksel olarak anlamlıdır.

## Örnek

X = [10, 20 ,30, 20, 25]

Y = [6, 8, 5, 2, 3] grupları arasındaki korelasyon katsaysını hesaplayalım. Daha sonra da P = 0 hipotezini test edelim. Az sayıda gözlem olduğu için dağılım hakkında yorum yapmamız mümkün değil. Bu yüzden Spearman'ın Sıra korelasyon katsayısı ölçütü kullanacağız.

### Spearman'ın sıra (rank) korelasyon katsayısı

```python
X = [10, 20 ,30, 20, 25]
Y = [6, 8, 5, 2, 3]
import scipy.stats
r, p = scipy.stats.spearmanr(X, Y)
print("Korelasyon katsayısı:",r, "/" ,"P değeri:",p)
```

+ Korelasyon katsayısı -0.30 çıktı. Yani X ve Y grupları arasında negatif yönde zayıf bir ilişkinin olduğunu söyleyebiliriz.

### P = 0'ın Anlamlılık Testi 

$H_0: P = 0$

$H_1: P \neq 0$

```python
#Test istatistiğinin hesabı
n = len(X) #Y de olabilirdi. ikisi de aynı gözlem değerine sahip 
test_stat = r / np.sqrt((1-r**2)/(n-2))
print(test_stat)
```

```python
#Tablo değerinin hesabı
a = 0.05
t_table_upper = scipy.stats.t.ppf(1-(a/2), df = n-2)
t_table_lower = scipy.stats.t.ppf(a/2, df = n-2)
print("Alt sınır:",t_table_lower, "/", "Üst sınır:",t_table_upper)
```

+ Test istatistiği ret bölgesine düştüğü için $H_0$ hipotezi reddedilir. Yani korelasyon katsayısı anlamlıdır.

## Örnek 

52 gözlemli bir örneklemde X ve Y grupları arasındaki korelasyon katsayısı 0.65 olarak bulunmuştur. $H_0: P = 0.85$ / $H_1: P > 0.85$ hipotezini a = 0.05'te test edelim.

```python
#Test istatistiğinin hesabı
import math
import numpy as np
n = 52
r = 0.65
p = 0.85
Z_r = 1/2*(math.log((1+r)/(1-r)))
Z_p = 1/2*(math.log((1+p)/(1-p)))
z_test_stat = (Z_r - Z_p)*np.sqrt(n-3)
print("Test istatistiği:",z_test_stat)
```

```python
#Tablo değerinin hesabı
import scipy.stats
a = 0.05
z_table_upper = scipy.stats.norm.ppf(1-a)
print(z_table_upper)
```

+ Görüleceği üzere test istatistiği kabul bölgesine düşüyor. $H_0$ hipotezi reddedilemiyor. Yani bulunan korelasyon katsayısı anlamsız.
