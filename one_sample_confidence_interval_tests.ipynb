## Güven Aralığı Formülleri

**Kitle Ortalaması İçin Güven Aralığı Formülleri:**

+ Kitle varyansı biliniyor: P($\bar{x}$ - $z_{\alpha/2}$ . $\frac{\sigma}{\sqrt{n}}$ $\leq$ $\mu$ $\leq$ $\bar{x}$ + $z_{\alpha/2}$ . $\frac{\sigma}{\sqrt{n}}$)


+ Kitle varyansı bilinmiyor + n $\geq$ 30: P($\bar{x}$ - $z_{\alpha/2}$ . $\frac{s}{\sqrt{n}}$ $\leq$ $\mu$ $\leq$ $\bar{x}$ + $z_{\alpha/2}$ . $\frac{s}{\sqrt{n}}$) 


+ Kitle varyansı bilinmiyor + n $\leq$ 30: P($\bar{x}$ - $t_{\frac{\alpha}{2}, n-1}$
 . $\frac{s}{\sqrt{n}}$ $\leq$ $\mu$ $\leq$ $\bar{x}$ + $t_{\frac{\alpha}{2}, n-1}$
 . $\frac{s}{\sqrt{n}}$) 
 
 
**Kitle Varyansı ve Standart Sapması İçin Güven Aralığı Formülleri:**

+ Varyans için güven aralığı: P($(n-1)\frac{s^2}{\chi^2_{1-\frac{\alpha}{2},n-1}}$ $\leq$ $\sigma^2$ $\leq$ $(n-1)\frac{s^2}{\chi^2_{\frac{\alpha}{2},n-1}}$)


+ Standart sapma için güven aralığı: P($s \sqrt{\frac{n-1}{\chi^2_{1-\frac{\alpha}{2},n-1}}}$ $\leq$ $\sigma$ $\leq$ $s \sqrt{\frac{n-1}{\chi^2_{\frac{\alpha}{2},n-1}}}$)


**Kitle Oranı İçin Güven Aralığı Formülü:**

+ Oran için güven aralığı: P($\hat{p} - z_{\frac{\alpha}{2}} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}} \leq p \leq \hat{p} + z_{\frac{\alpha}{2}} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$)

## Z, T ve Ki-Kare Tablolarından Tablo Değeri Hesabı

**NOT:**

+ Hesaplamaları göstermeden önce şu bilgiyi vermekte fayda var: Eğer ppf fonksiyonundaki 'q' parametresine gireceğimiz değeri '1-0.05' şeklinde yazarsak dağılımın sağ ucundaki pozitif değeri döndürür. Direkt olarak '0.05' şeklinde yazarsak dağılımın sol ucundaki negatif değeri döndürür.

**Z Tablosundan Tablo Değeri Bulma ($z_{\alpha/2}$):**

+ scipy.stats.norm = https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
import scipy.stats

#1-a = 0.99, a = 0.01 ve a/2 = 0.005 iken Z0.005 tablo değerini bulacağız.
z_score = scipy.stats.norm.ppf(q = 1-0.005, loc = 0, scale = 1)
print("%99 güven düzeyinde z skoru (tablo değeri) :", z_score)

#1-a = 0.95, a = 0.05 ve a/2 = 0.025 iken Z0.025 tablo değerini bulacağız
z_score_1 = scipy.stats.norm.ppf(q = 1 - 0.025, loc = 0, scale = 1)
print("%95 güven düzeyinde z skoru (tablo değeri) :", z_score_1)

#1-a = 0.90, a = 0.10 ve a/2 = 0.05 iken Z0.05 tablo değerini bulacağız
z_score_2 = scipy.stats.norm.ppf(q = 1-0.05, loc = 0, scale = 1)
print("%90 güven düzeyinde z skoru (tablo değeri) :", z_score_2)
```

**T Tablosundan Tablo Değeri Bulma ($t_{\frac{\alpha}{2}, n-1}$)**

+ scipy.stats.t = https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html#scipy.stats.t

```python
import scipy.stats

#1-a = 0.99, a = 0.01, a/2 = 0.005 ve df = 10 iken 
#T0.005,10 tablo değerini bulacağız.
t_score = scipy.stats.t.ppf(q = 1-0.005, df = 10, loc = 0, scale = 1)
print("%99 güven düzeyinde t skoru (tablo değeri) :", t_score)

#1-a = 0.95, a = 0.05, a/2 = 0.025 ve df = 10 iken 
#T0.025,10 tablo değerini bulacağız.
t_score_1 = scipy.stats.t.ppf(q = 1-0.025, df = 10, loc = 0, scale = 1)
print("%95 güven düzeyinde t skoru (tablo değeri) :", t_score_1)

#1-a = 0.90, a = 0.10, a/2 = 0.05 ve df = 10 iken 
#T0.05,10 tablo değerini bulacağız.
t_score_2 = scipy.stats.t.ppf(q = 1-0.05, df = 10, loc = 0, scale = 1)
print("%90 güven düzeyinde t skoru (tablo değeri) :", t_score_2)
```

+ t.ppf hesaplama yaparken, norm.ppf ile aynı mantığı kullanır. Sadece degrees of freedom (serbestlik derecesi = df) T dağılımında etkilidir ve bu fark yaratır. 

**Ki-Kare Tablosundan Tablo Değeri Bulma (${\chi^2_{1-\frac{\alpha}{2},n-1}}$ ve ${\chi^2_{\frac{\alpha}{2},n-1}}$)**

+ scipy.stats.chi2 = https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2.html

```python
import scipy.stats

#1-a = 99, a = 0.01, a/2 = 0.005 ve df = 9 iken
#Ki-Kare 1-a/2, n-1 (Upper) ve Ki-Kare a/2, n-1 (Lower) tablo değerlerini bulacağız
chi2_upper = scipy.stats.chi2.ppf(q = 1-0.005, df = 9)
chi2_lower = scipy.stats.chi2.ppf(q = 0.005, df = 9)
print("%99 güven düzeyinde chi2 tablo değerleri:", chi2_lower, chi2_upper)

#1-a = 95, a = 0.05, a/2 = 0.025 ve df = 9 iken
#Ki-Kare 1-a/2, n-1 (Upper) ve Ki-Kare a/2, n-1 (Lower) tablo değerlerini bulacağız
chi2_upper_1 = scipy.stats.chi2.ppf(q = 1 - 0.025, df = 9)
chi2_lower_1 = scipy.stats.chi2.ppf(q = 0.025, df = 9)
print("%95 güven düzeyinde chi2 tablo değerleri:", chi2_lower_1, chi2_upper_1)

#1-a = 90, a = 0.10, a/2 = 0.05 ve df = 9 iken
#Ki-Kare 1-a/2, n-1 (Upper) ve Ki-Kare a/2, n-1 (Lower) tablo değerlerini bulacağız
chi2_upper_2 = scipy.stats.chi2.ppf(q = 1-0.05, df = 9)
chi2_lower_2 = scipy.stats.chi2.ppf(q = 0.05, df = 9)
print("%90 güven düzeyinde chi2 tablo değerleri:", chi2_lower_2, chi2_upper_2)
```

### Örnek 1
Bir spor kulubündeki sporcuların boy uzunluklarının normal dağılıma sahip olduğu kabul ediliyor n= 50 olan bir örneklemden boy uzunluğunun ortalaması 165 ve standart sapması 25 bulunuyor Kitlenin boy uzunluğunun ortalaması için %90’lık güven aralığını tahmin ediniz

### Çözüm 
Örneklem Ortalaması ($\bar{x}$) = 165 

Örneklem Standart Sapma ($s^2$) = 25

Yanılma Olasılığı ($a$) = 0.10

Güven Düzeyi ($1-a$) = 0.90

Örneklem Boyutu ($n$) = 50

Uygulanacak Formül: P($\bar{x}$ - $z_{\alpha/2}$ . $\frac{s}{\sqrt{n}}$ $\leq$ $\mu$ $\leq$ $\bar{x}$ + $z_{\alpha/2}$ . $\frac{s}{\sqrt{n}}$)

```python
#1. Yol

#a = 0.10, a/2 = 0.05 iken za/2'yi buluyoruz
import scipy.stats
alpha = 0.10
z_table = scipy.stats.norm.ppf(q = 1-alpha/2, loc = 0, scale = 1)
print("%90 güven düzeyinde Z tablosunun tablo değeri:" ,z_table)

ayrac()

#Hesaplamayı yapıyoruz
# LCI = Left Confidence Interval, RCI = Right Confidence Interval
import numpy as np
x_bar = 165
s_kare = 25
n = 50
LCI = x_bar - z_table * (np.sqrt(s_kare)/(np.sqrt(n)))
RCI = x_bar + z_table * (np.sqrt(s_kare)/(np.sqrt(n)))
print("Kitle ortalaması için %90 güven aralığının alt ve üst sınırı:\n", LCI, RCI)
```

+ 1. yolda z tablosunun tablo değerini **scipy.stats.norm.ppf** ile bulduk. Daha sonra alt ve üst sınırları elimizle hesaplayacak şekilde **LCI ve RCI** değişkenlerine atadık.

```python
#2. Yol

#Direkt olarak güven aralığı hesaplayan bir işlev kullanacağız
confidence_interval = scipy.stats.norm.interval(confidence = 0.90, loc = x_bar, 
                                               scale = np.sqrt(s_kare)/(np.sqrt(n)))
print(confidence_interval)
```

+ 2. yolda **scipy.stats.norm.interval** işlevi ile direkt olarak güven aralığını hesapladık. **confidence** parametresine 0.90 güven düzeyini girdik, **loc** parametresine ortalamayı, **scale** parametresine $s/\sqrt{n}$ yazdık.

ortalaması için %95 güven düzeyindeki güven aralığını tahmin ediniz

### Çözüm
Örneklem Ortalaması ($\bar{x}$) = 0.4

```python
#1. Yol

#a = 0.05, a/2 = 0.025
alpha = 0.05
z_table = scipy.stats.norm.ppf(q = 1 - alpha/2, loc = 0, scale = 1)
print(z_table)
ayrac()

#LCI = Left Confidence Interval, RCI = Right Confidence Interval
x_bar = 0.4
sigma = 0.05
n = 25
LCI = x_bar - z_table * (sigma/np.sqrt(n))
RCI = x_bar + z_table * (sigma/np.sqrt(n))
print("Kitle ortalaması için %90 güven aralığının alt ve üst sınırı:\n", LCI, RCI)
```

### Örnek 3 
Organik yumurta satan bir markette bulunan 10’luk paketten rastgele bir örneklem alınıyor ve ağırlıkları gram olmak üzere; 58,60,65,60,63,70,55,67,58,69 olarak veriliyor. Kitle
varyansı ve standart sapması için %95 güven aralığını bulunuz

### Çözüm 
Yumurtalar ve Gramları = [58,60,65,60,63,70,55,67,58,69]

Güven Düzeyi ($1-a$) = 0.95

Yanılma Olasılığı ($a$) = 0.05

Kitle Varyansı İçin Kullanılacak Formül: P($(n-1)\frac{s^2}{\chi^2_{1-\frac{\alpha}{2},n-1}}$ $\leq$ $\sigma^2$ $\leq$ $(n-1)\frac{s^2}{\chi^2_{\frac{\alpha}{2},n-1}}$)

Kitle Standart Sapması İçin Kullanılacak Formül: P($s \sqrt{\frac{n-1}{\chi^2_{\frac{\alpha}{2},n-1}}}$ $\leq$ $\sigma$ $\leq$ $s \sqrt{\frac{n-1}{\chi^2_{1-\frac{\alpha}{2},n-1}}}$)

İlk olarak yumurataların varyanslarını ($s^2$) ve standart sapmalarını ($s$) bulmamız gerek.

```python
eggs = np.array([58,60,65,60,63,70,55,67,58,69])
s_kare = eggs.var()
s = eggs.std()
print("Yumurtaların varyansı :", s_kare)
print("Yumurtaların standart sapması :", s)
```

Ardından formülleri uygulmaya geçebiliriz. İlk olarak %95 güven düzeyinde kitle varyansı için güven aralığını bulalım.

```python
#1. Yol

#a = 0.05, a/2 = 0.025
alpha = 0.05
n = len(eggs)
chi2_table_upper = scipy.stats.chi2.ppf(q = 1-alpha/2, df = n-1)
chi2_table_lower = scipy.stats.chi2.ppf(q = alpha/2, df = n-1)
print("%95 güven düzeyinde chi2 tablo değerleri:", chi2_table_lower, chi2_table_upper)
ayrac()

##LCI = Left Confidence Interval, RCI = Right Confidence Interval
s_kare = 23.45
LCI = (n-1) * (s_kare/chi2_table_upper)
RCI = (n-1) * (s_kare/chi2_table_lower)
print("Kitle varyansı için %95 güven düzeyinde, güven aralığı:\n", LCI, RCI)
```

Şimdi de %95 güven güzeyinde kitle standart sapması için güven aralığını bulalım.

```python
#1. Yol

print("s =", s, "\n"
      "n =", n, "\n"
     "chi2_table_upper =", chi2_table_upper,"\n"
     "chi2_table_lower =", chi2_table_lower)
ayrac()

#LCI = Left Confidence Interval, RCI = Right Confidence Interval
LCI = s * np.sqrt(n-1/chi2_table_lower)
RCI = s * np.sqrt(n-1/chi2_table_upper)
print("Kitle standart sapması için %95 güven düzeyinde, güven aralığı:\n", LCI, RCI)
```

Kitle ortalaması için %95 güven aralığını oluşturunuz .

### Çözüm

Örneklem Ortalaması ($\bar{x}$) = 50

Örneklem Standart Sapması ($s$) = 8

Örneklem Büyüklüğü ($n$) = 25

Güven Düzeyi ($1-a$) = 0.95

Uygulanacak Formül: P($\bar{x}$ - $t_{\frac{\alpha}{2}, n-1}$

### Örnek 5
Bir bölgede bir gün içinde PCR test yapılan 500 kişinin 175’inde
test sonucu pozitif çıkmıştır . Sonucu pozitif olanların kitle oranı için %98
güven düzeyinde güven aralığını hesaplayınız

### Çözüm
Örneklem Büyüklüğü ($n$) = 500

Pozitif Çıkanlar ($+$) = 175

Pozitif Çıkanların Oranı ($\hat{p}$) = $175/500$ = 0.35

Güven Düzeyi ($1-a$) = 0.98

Yanılma Olasılığı ($a$) = 0.02

Uygulanacak Formül : P($\hat{p} - z_{\frac{\alpha}{2}} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}} \leq p \leq \hat{p} + z_{\frac{\alpha}{2}} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$)

```python
#1. Yol

#a = 0.02, a/2 = 0.01
alpha = 0.02
z_table = scipy.stats.norm.ppf(q = 1-alpha/2, loc = 0, scale = 1)
print(z_table)

ayrac()

#LCI = Left Confidence Interval, RCI = Right Confidence Interval
p_hat = 0.35
n = 500
LCI = p_hat - z_table * np.sqrt(p_hat* (1 - p_hat)/n)
RCI = p_hat + z_table * np.sqrt(p_hat* (1 - p_hat)/n)
print("%98 güven düzeyinde kitle oranı için güven aralığı: \n", LCI, RCI)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Veri setini yükleme
iris = load_iris()

# Özellikleri X'e, hedef değişkeni y'ye atama
X = iris.data
y = iris.target

# Eğitim ve test verilerini ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modeli oluşturma
knn = KNeighborsClassifier(n_neighbors=3)

# Modeli eğitme
knn.fit(X_train, y_train)

# Test verilerini kullanarak doğruluk oranını hesaplama
accuracy = knn.score(X_test, y_test)

print("Doğruluk oranı:", accuracy)
```

Bir istatistikçi bir otelin odalarının doluluk durumunu incelemek amacıyla rasgele olarak

standandart sapmasını 5,86 olarak hesaplıyor. Kiralanan oda sayısının dağılımının normal
dağılıma uyduğu varsayımı altında, kitle standart sapması için %95 güven düzeyinde bir
aralık tahmini yapınız.

## Çözüm

Örneklemdeki Öğe Sayısı ($n$) = 27

Standart Sapma ($s$) = 5.86

Uygulanacak Formül: P($s \sqrt{\frac{n-1}{\chi^2_{\frac{\alpha}{2},n-1}}}$ $\leq$ $\sigma$ $\leq$ $s \sqrt{\frac{n-1}{\chi^2_{1-\frac{\alpha}{2},n-1}}}$)

Yanılma Olasılığı ($a$) = 0.05

```python
import scipy.stats
import numpy as np

a = 0.05
n = 27
s = 5.86

#Tablo değerlerinin bulunması
chi2_table_upper = scipy.stats.chi2.ppf(q = 1-(a/2), df = n-1) 
chi2_table_lower = scipy.stats.chi2.ppf(q = a/2, df = n-1)
print(chi2_table_upper)
ayrac()
print(chi2_table_lower)
ayrac()

#Formülün uygulanması
ust_sınır = s*(np.sqrt((n-1)/chi2_table_lower))
alt_sınır = s*(np.sqrt((n-1)/chi2_table_upper))

#Sonuç
print("Kitle standart sapmasının %95 güven düzeyinde güven aralığı:",alt_sınır,",",ust_sınır)
```
